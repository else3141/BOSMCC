#Task Description  
Accept input as a local file and return a numpy array containing a 5 x [No. of frames] matrix. The matrix should contain a row for each frame of the video. The columns are as follows:  
[player fallen in frame or not, x coordinate, y coordinate, width of player, height of player]  

Note that x,y,w,h must be integral values  

Do not rescale the video. Outputs will be checked using RMSE, assuming a resolution of [Resolution]  







```python
from ultralytics import YOLO

model = YOLO("yolov8m.pt")
result = model.predict("manc.mp4", save = True)
```

```python
print(result[0:192])
```

```python
frames = []
```

```python
for  j in range(192):
    frames.append([0,0,0,0,0])
    for i in result[j].boxes:
        if list(i.cls) == [0.]:
            x,y,w,h = list(i.xywh)[0]
            if w>1.5*h:
                frames[-1][0] = 1
                x = int(x)
                y = int(y)
                w = int(w)
                h = int(h)
                frames[-1][1:] = [x,y,w,h]
                print(frames[-1])
```

```python
import numpy as np
```

```python
fin = np.matrix(frames)
```
